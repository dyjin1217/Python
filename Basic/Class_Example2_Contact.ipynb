```python
class Contact:
    def __init__(self, name, phone_number, e_mail, addr):
        self.name = name
        self.phone_number = phone_number
        self.e_mail = e_mail
        self.addr = addr

    def print_info(self):
        print("Name: ", self.name)
        print("Phone Number: ", self.phone_number)
        print("E-mail: ", self.e_mail)
        print("Address: ", self.addr)
        
    def get_name(self):
        return self.name

class ContactHandler:
    def __init__(self):
       self.contact_list = []
                    
    def set_contact(self):
        name = input("Name: ")
        phone_number = input("Phone Number: ")
        e_mail = input("E-mail: ")
        addr = input("Address: ")
        contact = Contact(name, phone_number, e_mail, addr)
        self.contact_list.append(contact)

    def print_contact(self):
        for contact in self.contact_list:
            contact.print_info()
                
    def delete_contact(self):
        name = input("Name: ")
        for i, contact in enumerate(self.contact_list):
            if contact.get_name() == name:
                del self.contact_list[i]
        
def print_menu():
    print("1. 연락처 입력")
    print("2. 연락처 출력")
    print("3. 연락처 삭제")
    print("4. 종료")
    menu = input("메뉴선택: ")
    return int(menu)

def main():
    h = ContactHandler()
    
    while 1:
        menu = print_menu()
        if menu == 1:
            h.set_contact()
        elif menu == 2:
            h.print_contact()
        elif menu == 3:
            h.delete_contact()
        elif menu == 4:
            break

main()
```
